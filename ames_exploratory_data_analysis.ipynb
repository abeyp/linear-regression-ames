### SalePrice Trends over Time?

Time Series analysis showing SalePrice over the given Month and Year Sold data. 

In the Ames dataset, we are given a `MoSold` and `YrSold` as the only indicators of measurement for time series analysis. 

```python
data[['SalePrice','MoSold','YrSold']].sort_values(by='YrSold')
```

**Need to create a DateTime Pandas Column/Series to work with for plotting a visualization**

```python
# Create default day column to help with creating datetime

data = data.assign(DaySold = '1')
```

```python
data[['SalePrice','DaySold','MoSold','YrSold']].sort_values(by='YrSold')
```

```python
# Rename columns to fit my `.to_datetime` method

data.rename(columns={'YrSold':'Year', 'MoSold':'Month','DaySold':'Day'}, inplace=True)
data.head()
```

```python
# Create our Datetime Series

data['Datetime'] = pd.to_datetime(data[['Year', 'Month', 'Day']])
```

```python
#Check our our DateTime column against the year, month, day

data[['Datetime', 'SalePrice', 'Year', 'Month','Day']]
```

Now that our DateTime is set, we can provide better visualizations of SalePrice trends over time. If we were to currently stop here and start plotting our points, our best bet is creating a scatter plot to see every house datapoint.

```python
plt.scatter('Month', 'SalePrice', data=data)
plt.xticks(rotation = 45)
plt.legend()
```

For this analysis, we will use Summary statistics to inform the trend over time of SalePrice.

```python
#Set a datetime index

data.set_index('Datetime',inplace=True)
```

```python
# Keep a datetime column for upcoming .describe calculations

data["DateTime_col"] = data.index
```

Next, we can use `.describe()` to compute the SalePrice mean, std, and median. 

Creating a new dataframe, `time_df` for our summary statistics to be mapped back to the Datetime in`data`.

```python
time_df = data[['SalePrice','DateTime_col']].groupby('DateTime_col').describe()
time_df
```

time_df uses a MultiIndex `['SalePrice','count']`, we will need to access the Level 1 where our summary statistic columns lie.

```python
# checking that we are pulling values needed

time_df.loc["2007-01-01", time_df.columns.get_level_values(1)=='mean'][0]
```

Helper functions to return the summary statistics values needed for each month

```python
def populate_median(s):
    s = str(s).split(" ")[0]
    return time_df.loc[s, time_df.columns.get_level_values(1)=='50%'][0]

        
def populate_std(s):
    s = str(s).split(" ")[0]
    try:
        return time_df.loc[s, time_df.columns.get_level_values(1)=='std'][0]
    except:
        print("bruh something happened std", s)
        
def populate_mean(s):
    s = str(s).split(" ")[0]
    try:
        return time_df.loc[s, time_df.columns.get_level_values(1)=='mean'][0]
    except:
        print("bruh something happened mean", s)
```

Here we are mapping our summary statistics as new columns in `data`

```python
data["monthly_median"] = data["DateTime_col"].apply(populate_median)
data["monthly_std"] = data["DateTime_col"].apply(populate_std)
data["monthly_mean"] = data["DateTime_col"].apply(populate_mean)

#df_data["monthly_mean"].head(50)
#df_data["monthly_std"].head(50)
#df_data["monthly_mean"].head(50)
```

Finally here are the summary statistics columns now in our `data`

```python
data[['monthly_mean','monthly_std','monthly_median']]
```

Now starting to create our vizualization, we will sort our index for the plot.

```python
data.sort_index(inplace=True)
```

```python
plt.plot(data.index, 'monthly_median', data=data)
plt.plot(data.index, 'monthly_mean', data=data)
# plt.plot(data.index, 'monthly_std', data=data)

plt.title("SalePrice Trends over Time (MM-YY)")
plt.ylabel('SalePrice')
plt.xlabel('MM-YY')

plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.xticks(rotation = 45)

plt.legend()
plt.show
```

Now we have a concise line graph with great interpretiability compared to the beginning scatter plot. 

SalePrice against time with labels indicating our monthly mean and monthly median for all homes.

Taking a first glance at our line graph, there are cyclical trends present and the well-known 2008 event shown with the drastic increase and decrease in SalePrice for this neighborhood.
